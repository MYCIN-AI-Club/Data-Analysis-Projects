**LINEAR REGRESSION RELATED TO ABOARD**

```python
data.plot(x="Aboard", y="Fatalities", kind="scatter")
plt.show()
```

```python
def std_based(col_name,df):
    mean = df[col_name].mean()
    std = df[col_name].std()
    cut_off = std * 2
    lower, upper = mean - cut_off, mean + cut_off
    new_df = df[(df[col_name] < upper) & (df[col_name] > lower)]
    return new_df
```

**REMOVING OUTLIERS OF FATALITIES COLUMN :**

```python
fatality_cols = ['Fatalities']
```

```python
print(f"dataset size before removing outliers= {data.shape[0]}")
for col in fatality_cols:
    data = std_based(col, data)
print(f"dataset size after removing outliers= {data.shape[0]}")
```

**REMOVING OUTLIERS OF ABOARD COLUMN :**

```python
abroad_cols = ['Aboard']
```

```python
print(f"dataset size before removing outliers= {data.shape[0]}")
for col in abroad_cols:
    data = std_based(col, data)
print(f"dataset size after removing outliers= {data.shape[0]}")
```
